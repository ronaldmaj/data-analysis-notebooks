I extended this dataset by a factor of 4 by creating transformations of the orignals including rotations, sharpenings and adjustments to brightness.

Training the model using the imageai module (with a ResNet setting) provided an accuracy of over 90% on the test set, however using a sliding window search on an example image has left more to be desired.

Introducing a third category (partial-ship) did not increase the ability of the model to pick up ship images.

### Scraping Container Data

Before finalizing the model above, I have also scraped the data on the number of TEUs (Twenty-foot Equivalent Units) or shipping containers from the Port of Los Angeles website:
- https://www.portoflosangeles.org/business/statistics/container-statistics/historical-teu-statistics-2018

Using Beautiful soup I scraped the website and using Pandas I stored that data and did some initial analysis, plotting the imports, exports and total number of containers for a number of recent years (2012-2018)

The changes in export and import volumes can be seen for the last few years in the figures below.

##### Notebooks:
The notebook that details the scraping process:
- https://github.com/ronaldmaj/data-analysis-notebooks/blob/master/POLA_Project_4_-_Scraping_Container_Data.ipynb

```python
import pandas as pd
import datetime

font = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : 14}
plt.style.use('ggplot')
plt.rc('font', **font)

df_cont = pd.read_csv('Container_Data_1996-2018.csv')

def plot_yr(yr):
    plt.figure(figsize=(9,6))
    months = [
        datetime.date(yr,int(mn+1),1).strftime('%b')
        for mn in df_cont.loc[df_cont['Year']==yr,'Month']
    ]
    plt.plot(months,
             df_cont.loc[df_cont['Year']==yr,'Total TEUs'],
             label='Total TEUs')
    
    plt.plot(months,
             df_cont.loc[df_cont['Year']==yr,'Total Imports'],
             label='Total Imports')
    
    plt.plot(months,
             df_cont.loc[df_cont['Year']==yr,'Total Exports'],
             label='Total Exports')
    
    plt.title(str(yr))
    plt.ylabel('TEUs (Containers)')
    plt.xlabel('Month')
    plt.legend()
    plt.show()


for i in range(2012,2019):
    plot_yr(i)
```

Although the recent tariffs introduced in 2018 were expected to decrease trade in/out of the USA, it does not seem like this has been the case for container units in the Port of Los Angeles. 

This may be due to a delay in the effect occuring or perhaps other ports have seen more of the impact thus far. Further analysis is necessary to determine the reason

Trends include imports exceeding exports every year and almost every month.

However, Feb - Mar seems to be a generally low trade month with both imports and exports down. 
<br> This is also the only time where US exports occationally exceed imports
